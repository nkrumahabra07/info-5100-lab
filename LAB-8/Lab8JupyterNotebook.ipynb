## Box Plots

````python
plt.figure(figsize=(8, 6))
sns.boxplot(x="Species", y="SepalLengthCm", data=iris_dataframe, palette="Set3")

# Add labels and title
plt.title("Box Plot of Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.show()
````

## Swarm Plot + Box Plot

````python
plt.figure(figsize=(8, 6))
sns.boxplot(x="Species", y="SepalLengthCm", data=iris_dataframe, palette="pastel")
sns.swarmplot(x="Species", y="SepalLengthCm", data=iris_dataframe, color=".25")
plt.title("Box and Swarm Plot of Petal Width by Species")
plt.xlabel("Species")
plt.ylabel("Petal Width (cm)")
plt.show()
````

## Kernel Density Plot (KDE)

KDE plot helps to visualize the distribution of data in a dataset



````python
# Create a KDE plot
plt.figure(figsize=(8, 6))
sns.kdeplot(data=Bakerydata_dataframe, x='Transaction', hue='period_day', fill=True, common_norm=False, palette="crest", alpha=0.2)

# Add labels and title
plt.title("Kernel Density Plot of Sepal Length by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel('period_day')
plt.show()
````

## Violin Plot (Box Plot + KDE)

A statistical graphic that shows the distribution of numerical data by combining a box plot and a kernel density plot.

````python
sns.violinplot(x='Species', y='PetalWidthCm', data=iris_dataframe)
plt.title('Violin Plot of Petal Width by Species')
plt.show()
````

## Correlation Heat Maps

````python
correlation = iris_dataframe.iloc[:, :-1].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correl ation Heatmap')
plt.show()
````
